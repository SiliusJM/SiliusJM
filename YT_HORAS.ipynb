```python
import yt_dlp
import pandas as pd

playlist_url = "https://www.youtube.com/playlist?list=PLVQkt0PdAIQGrLmihYrHj4Ubri0lz06tN"

ydl_opts = {
    "quiet": True,
    "skip_download": True
}

videos = []

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    info = ydl.extract_info(playlist_url, download=False)

    for i, entry in enumerate(info["entries"], start=1):
        title = entry.get("title", "N/A")
        duration = entry.get("duration", 0)  # en segundos
        videos.append({
            "NÂ°": i,
            "Nombre": title,
            "Duracion_seg": duration
        })

df = pd.DataFrame(videos)

# ğŸ”¢ DuraciÃ³n total
total_segundos = df["Duracion_seg"].sum()

horas = total_segundos // 3600
minutos = (total_segundos % 3600) // 60

print(f"â±ï¸ DuraciÃ³n total de la playlist: {horas} horas y {minutos} minutos")

# ğŸ“Œ FunciÃ³n para dividir en partes
def puntos_de_corte(df, partes):
    objetivo = total_segundos / partes
    acumulado = 0
    cortes = []

    for _, row in df.iterrows():
        acumulado += row["Duracion_seg"]
        if acumulado >= objetivo and len(cortes) < partes - 1:
            cortes.append((row["NÂ°"], row["Nombre"]))
            objetivo += total_segundos / partes

    return cortes

# Mostrar cortes
for partes in [2, 3, 4]:
    print(f"\nğŸ“º Para ver en {partes} partes, continÃºa desde:")
    cortes = puntos_de_corte(df, partes)
    for i, (num, nombre) in enumerate(cortes, start=2):
        print(f"  Parte {i}: Video #{num} â†’ {nombre}")


```
